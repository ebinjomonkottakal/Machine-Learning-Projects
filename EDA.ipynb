# EDA using Univariate Analysis

```python
import pandas as pd
import seaborn as sns
titanic = pd.read_csv('train.csv')
```

```python
titanic.head(5)
```

## 1.Categorical Data


#### a.countplot

```python
sns.countplot(titanic['Survived'])
titanic['Survived'].value_counts()
```

```python
sns.countplot(titanic['Pclass'])
```

```python
sns.countplot(titanic['Sex'])
```

#### b.Piechart

```python
titanic['Survived'].value_counts().plot(kind = 'pie',autopct = '%.2f')
```

```python
sns.displot(titanic['Age'])
```

### c. Boxplot

```python
sns.boxplot(titanic['Age'])
```

## EDA using Bivariate Analysis and Multivariate Analysis

```python
tip = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')

sns.set(rc = {'figure.figsize':(11,6)})
```

### a. Scatter plot (Numerical - Numerical)

```python
tip.head()
```

```python
flights.head()
```

```python
sns.scatterplot(tip['total_bill'],tip['tip'],hue = tip['sex'],style = tip['smoker'],size = tip['size'])
```

### b. Bar Plot (Numerical-Category)

```python
sns.barplot(titanic['Pclass'],titanic['Fare'],hue = titanic['Sex'])
```

### c. Box Plot (Numerical - Categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue = titanic['Survived'])
```

### d. Dist Plot

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist = False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist = False)
```
