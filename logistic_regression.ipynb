```python
from sklearn.datasets import make_classification
```

```python
import numpy as np
import pandas as pd
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, 
                    n_redundant=0, n_classes=2, n_clusters_per_class=1,
                          class_sep=10,random_state = 54,hypercube=False)
X = np.insert(X,0,1,axis = 1)
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize =(10,6) )
plt.scatter(X[:,1],X[:,2],c=y,cmap = 'winter',s=100)
```

```python
def perceptron(x,y):
  x = X
  w = np.ones(X.shape[1])
  epoch = 2000
  coef  = 0.01

  for i in range(epoch):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],w))
    w = w+ coef*(y[j]-y_hat)*X[j]
  return w[0],w[1:]
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
intercept,coeff = perceptron(X,y)
```

```python
intercept,coeff
```

```python
m = coeff[0]/coeff[1]
c = intercept/coeff[1]
```

```python
x_input = np.linspace(-3,3,100)
y_input = (m*x_input)+c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color = 'red',linewidth = 3)
plt.scatter(X[:,1],X[:,2],c=y,cmap = 'winter',s=100)
```
